```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
prices = pd.read_csv('Prices_Gua2.csv')
hhchar = pd.read_csv('HH_Char.csv')
```

```python
prices.reset_index(inplace=True)
```

```python
prices.columns
```

```python
prices
```

```python
# separating household characteristics by region given in original spreadsheet, mainly to inspect the Rural variable for each

central = hhchar[hhchar['m']=='Central']
metropolitana = hhchar[hhchar['m']=='Metropolitana']
noroccidente = hhchar[hhchar['m']=='Noroccidente']
nororiente = hhchar[hhchar['m']=='Nororiente']
norte = hhchar[hhchar['m']=='Norte']
peten = hhchar[hhchar['m']=='Peten']
suroriente = hhchar[hhchar['m']=='Suroriente']
suroccidente = hhchar[hhchar['m']=='Suroccidente']

regions = [central, metropolitana, noroccidente, nororiente, norte, peten, suroriente, suroccidente]
```

```python
for region in regions: 
    plt.figure()
    #plt.title(f'{str(i['m'])} Urban-Rural Breakdown'.format(str)) 
    plt.hist(region['Rural'])
```

regions = central, metropolitana, noroccidente, nororiente, norte, peten, suroriente, suroccidente

so, we find

Rough Ratios, Urban:Rural

- Central: 1:2 (majority rural)
- Metropolitana: 8:1 (vast majority urban)
- Noroccidente: 1:2 (majority rural)
- nororiente: 2:1 (majority urban)
- Norte: 1:1 (even split)
- Peten: 0.8:1 (almost even split)
- Suroriente: 0.6:1 (majority rural)
- Suroccidente: 1:2 (majority rural)



```python
# plot prices for each region

region_str = ['Central', 'Metropolitana', 'Noroccidente', 'Nororiente',
           'Norte', 'Peten', 'Suroccidente', 'Suroriente']

for region in region_str:
    plt.figure(figsize = (10,10))
    plt.scatter(prices['j'], prices[region])
    plt.title(region)
```
